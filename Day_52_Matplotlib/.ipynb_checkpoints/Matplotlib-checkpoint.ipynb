# Matplotlib

1. Matplotlib is low level graph plotting library in python that serves as a visualization utility
2. Matplotlib is open source and we can use it 
3. Matplotlib is mostly written in python, few segement are written in c, Objective- c and Javscript for platform compatible
4. To install - [pip install matplotlib]

```python
import matplotlib
```

### Checking Version

```python
import matplotlib

print(matplotlib.__version__)
```

### Pyplot

1. Most of the Matplotlib utility lies under the pyplot submodule
2. We are usually under the plt.alias

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([0, 6])
y_axis = np.array([0,300])

plt.plot(x_axis, y_axis)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([2, 6])
y_axis = np.array([40,300])

plt.plot(x_axis, y_axis)
plt.show()
```

### Plotting x and y points

1. The plot() function is used to draw points (markers) in a diagram
2. By Default plot() function draws a line from point to point 
3. Plot function takes parameters for specifying points in Diagram:
    1. Parameter 1 is array containing the points on the x - axis [horizantal axis]
    2.  Parameter 2 is array containing the points on the y - axis [Vertical axis ]
4. Example - x[2, 6] any y[40,300]

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([1, 9])
y_axis = np.array([3,10])

plt.plot(x_axis, y_axis)
plt.show()
```

### Plotting without Line

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([1, 9])
y_axis = np.array([3,10])

plt.plot(x_axis, y_axis,'o')
plt.show()
```

### Multiple Points 

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([1, 9, 6, 10,15])
y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(x_axis, y_axis)
plt.show()
```

### Default Data 

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis)
plt.show()
```

### Markers

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, marker = '*')
plt.show()
```

### Format Strings [ fmt ]

1. This parameter is also called as fmt
2. Syntax written of fmt use like below [marker |line | color]
3. shortcut string notation used to specify parameters to specify markers 

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, 'o-')
plt.show()
```

### Line References

1. Solid line  [ '-' ]
2. Dotted Line [ ':' ]
3. Dashed Line [ '--']
4. Dashed Dotted ['-.']

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, 'o-r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, 'o--r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, 'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_axis = np.array([3, 10, 2, 5, 7])

plt.plot(y_axis, 'o-.r')
plt.show()
```
