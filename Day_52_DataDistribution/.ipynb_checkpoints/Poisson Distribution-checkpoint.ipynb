# 1. Poisson Distribution

1. Poisson Distribution also called Discrete Distribution
2. It estimate how many times an event can happen in a specified time.
3. Example:
    1. If someone eats twice a day what is probability he will eat a thrice?
4. It has two parameters:
    1. lam - rate of known number of occurances Example. 2 for above problem
    2. size - The shape of the returned array

```python
from numpy import random

x = random.poisson(lam =2, size =100)
print(x)
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(random.poisson(lam =2, size = 1000), kde = True)
plt.show()
```

```python
sns.displot(random.poisson(lam =2, size = 1000), kde = False)
plt.show()
```

### Difference between Normal and Poisson Distribution

```python
sns.distplot(random.normal(loc =50, scale =7, size =1000), hist=False, label='normal')
sns.distplot(random.poisson(lam =50, size =1000), hist = False, label = "Poisson")

plt.legend()
plt.show()
```

# 4. Chi Square Distribution

1. chi square distribution is used as a basis to verify hypothesis
2. It has two parameters:
    1. df - (degree of freedom)
    2. size - The shape of the returned array

```python
from numpy import random

x =  random.chisquare(df =2, size =(2,3))
print(x)
```

```python
from numpy import random

x =  random.chisquare(df =10, size =(2,3))
print(x)
```

# 5. Multinomial Distribution

1. Multinomial distribution is a generalization of binomial distribution
2. It describes the outcome of multinomal events whereas binomaial have only one of two events
3. It has three parameters:
    1. n - number of possible outcomes Example 6 for dice roll
    2. pvals - list of probability of outcomes Example 1/6, 1/6, 1/6, 1/6, 1/6,1/6 for dice roll
    3. size - The shaped of returned array

```python
x = random.multinomial(n =6, pvals = [1/6, 1/6, 1/6, 1/6, 1/6,1/6])
print(x)
```
