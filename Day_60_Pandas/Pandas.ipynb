## Replace Value using

1. Mean (Average):
    1. Average value (sum of all values divide by number of values(counts))
2. Median:
    1. The value in middle of dataset[feature], after you have sorted all values ascending
3. Mode:
    1. The most appear value in the dataset[feature]

## 1. Mean

```python
import pandas as pd

df = pd.read_csv('car_sales_data.csv')

x = df['price'].mean()

df['price'].fillna(x, inplace = True)
```

# Data Correlation

1. Pandas Module corr()
2. The corr() method calculate relationship between each column[features] in your datasets
3. The corr() method ignores 'Non Numeric' columns
4. The Number(Result) varies from -1 to 1
5. The Result of corr() method is a table with a lot of numbers that represent, how strong relationship between two columns in Dataset
6. There are three types of Relationship Result:
    1. Perfect Relationship
    2. Good Relatinship
    3. Bad Relationship

## 1. Perfect Relationship

1. 1 means that there is 1 to 1 relationship( a perfect Relationship)

## 2. Good Relationship

1. 0.9 is also good relationship, if you increase value [0.6 to close 1]
2. -0.9 is also good relationship, if you increase value [0.6 to close 1]

## 3.  Bad Relationship

1. Value less than 0.5

## What is Good Relationship?

Atleast (0.6) or (-0.6) above it is a good relationship
