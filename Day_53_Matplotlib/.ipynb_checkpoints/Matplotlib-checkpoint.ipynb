# Color References

1. Red ['r']
2. Green ['g']
3. Blue ['b']
4. Cyan ['c']
5. Megenta ['m']
6. Yellow ['y']
7. Black ['k']
8. White ['w']

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import sys
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, y_axis, 'o:g')
plt.show
```

# Marker Size & Color

1. Use markersize keyword or use shorter form [ms]
2. use markerfacecolor keyword argument or shorter form [mfc]

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, y_axis, 'o:g', ms = 10)
plt.show
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, y_axis, 'o:g', ms = 10, mec = 'r')
plt.show
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, y_axis, 'o:g', ms = 10, mec = 'r', mfc = 'r')
plt.show
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, y_axis, 'o:b', ms = 15, mfc = 'r')
plt.show
```

## Line Width [linewidth or lw]

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
matplotlib.use("Agg")

plt.plot(x_axis, ms = 10, mec = 'hotpink', mfc = 'hotpink',c = 'black', linewidth = '5')
plt.plot(y_axis)
plt.show
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])
z_axis = np.array([10,2,5,8,9,7,6])
matplotlib.use("Agg")

plt.plot(x_axis, ms = 10, mec = 'hotpink', mfc = 'hotpink',c = 'black', linewidth = '5')
plt.plot(y_axis)
plt.plot(z_axis)
plt.show
```

# Title and Labels

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])

plt.plot(x_axis,y_axis, ms = 10, mec = 'hotpink', mfc = 'hotpink',c = 'black', linewidth = '5')

plt.xlabel("Average Pulse")
plt.ylabel("Calories Burn")

plt.show
```

```python
x_axis = np.array([2,9,7,6,15,10,1])
y_axis = np.array([5,4,9,2,9,5,8])

plt.plot(x_axis,y_axis, ms = 10, mec = 'hotpink', mfc = 'hotpink',c = 'black', linewidth = '5')

plt.title("2022 Sport Data ")
plt.xlabel("Average Pulse")
plt.ylabel("Calories Burn")

plt.show
```

# Display Multiple Graphs

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,20,330])

plt.subplot(2,1,1)
plt.plot(x, y)


# plot 2
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,20,330])

plt.subplot(2,1,2)
plt.plot(x, y)

plt.show()
```
