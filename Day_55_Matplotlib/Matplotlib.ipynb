## Histogram

1. A histogram is a graph showing frequency distributions.
2. It is a graph showing the number of obeservation within each given interval
3. Example:
    1. Height of 250 people, you migh end up with a histogram like this
    
4. we use hist() function to create histogram
5. The hist() function will use an array of numbers to create histogram, array is sent into the function as an argument.

```python
import numpy as np

x = np.random.normal(170, 10,200)
print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10,200)

plt.hist(x)
plt.show()
```

## Explode

If I want one of the wedge stand out ?
1. Explode parameter allows you to do that
2. explode parameter, if specified and not None, must be an array with one of the value for each wedge


```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([25,35,45,55])
my_labels = ["apple","mango","cherry","banana"]
my_explode = [0.1, 0.1, 0.1, 0.1]

plt.pie(x_axis, labels = my_labels, explode = my_explode)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([25,35,45,55])
my_labels = ["apple","mango","cherry","banana"]
my_explode = [0, 0, 0, 0.1]

plt.pie(x_axis, labels = my_labels, explode = my_explode)
plt.show()
```

## Colors

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([25,35,45,55])
my_labels = ["apple","mango","cherry","banana"]
my_explode = [0, 0, 0, 0.1]
my_colors = ["red","black",'green','hotpink']

plt.pie(x_axis, labels = my_labels, explode = my_explode, shadow=True, colors = my_colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_axis = np.array([25,35,45,55])
my_labels = ["apple","mango","cherry","banana"]
my_explode = [0, 0, 0, 0.1]
my_colors = ["red","black",'green','hotpink']

plt.pie(x_axis, labels = my_labels, explode = my_explode, shadow=True, colors = my_colors)
plt.legend(title ="Fruits Data")
plt.show()
```

## Boxplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(10)
data = np.random.normal(100,20,200)

fig = plt.figure(figsize =(10, 7))

plt.boxplot(data)
plt.show()
```

```python
np.random.seed(10)
data_1 = np.random.normal(100,20,200)
data_2 = np.random.normal(90,20,200)
data_3 = np.random.normal(80,30,200)
data_4 = np.random.normal(70,40,200)

data = [data_1, data_2, data_3, data_4]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_axes([0, 0, 1, 1])

ax.boxplot(data)
plt.show()
```
